```python
import os
from pandas import read_csv
from base64 import b64decode
import re

def split_in_words(in_string):
    return [r.lower() for r in re.split('[\W\d_]', in_string) if r]

all_words_count_map = {}
for filename in os.listdir('.'):
    if filename.endswith("-method-data.csv"): 
        print(filename)
        data_frame = read_csv(filename, delimiter=';', header=0, keep_default_na=False)
        data_frame['checkinComment'] = data_frame['checkinComment'].apply(lambda s: b64decode(s).decode('utf-8'))
        for checkin_comment in data_frame['checkinComment']:
            comment_words_processed = []
            for w in split_in_words(checkin_comment):
                if w in all_words_count_map:
                    all_words_count_map[w]['occurs'] = all_words_count_map[w]['occurs'] + 1
                    if w not in comment_words_processed:
                        all_words_count_map[w]['commits'] = all_words_count_map[w]['commits'] + 1
                else:
                    all_words_count_map[w] = {'occurs': 1, 'commits': 1}
                comment_words_processed.append( w )
```

```python
from pandas import concat

wlist = ['fix', 'security', 'fails', 'fixed', 'error', 'missing', 'bug', 'failed', 'exception', 'fixes', 'memory', 'npe']
dframe_list = []
for filename in os.listdir('.'):
    if filename.endswith("-method-data.csv"): 
        print(filename)
        data_frame = read_csv(filename, delimiter=';', header=0, keep_default_na=False)
        print(len(data_frame))
        data_frame['is_interessting'] = data_frame['checkinComment'].apply(lambda s: any( w in b64decode(s).decode('utf-8') for w in wlist))
        data_frame = data_frame[data_frame['is_interessting']]
        del data_frame['is_interessting']
        print(len(data_frame))
        dframe_list.append(data_frame)
        data_frame.to_csv('interesting_comment/' + filename[:-4]+'-i.csv', index=False, sep=';')
data_frame = concat(dframe_list)
data_frame.to_csv('interesting_comment/all-interesting.csv', index=False, sep=';')
```
